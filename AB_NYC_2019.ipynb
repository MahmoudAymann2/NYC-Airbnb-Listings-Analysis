```python
Q1 = Q3 = np.quantile(df.minimum_nights, .25)
Q3 = np.quantile(df.minimum_nights, .75)
IQR = Q3 - Q1
min_upper = Q3 + 1.5 * IQR
```

```python
upper_outliers = df[df.minimum_nights > min_upper].minimum_nights.values
df.minimum_nights.replace(upper_outliers, min_upper, inplace = True)
```

```python
df.describe()
```

```python
Q1 = np.quantile(df["calculated_host_listings_count"], .25)
Q3 = np.quantile(df["calculated_host_listings_count"], .75)
IQR = Q3 - Q1
min_upper = Q3 +(1.5 * IQR)
min_upper = np.round(min_upper)
```

```python
upper_outliers = df[df["calculated_host_listings_count"] > min_upper].calculated_host_listings_count.values
df["calculated_host_listings_count"].replace(upper_outliers, min_upper, inplace = True)
```
